```python
# Import library

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
from scipy.signal import spectrogram

# Load data

path_major = '/Users/woojaejeong/Desktop/Program/USC/Computational Introduction to Deep Learning/Project/Data/Major/'
path_minor = '/Users/woojaejeong/Desktop/Program/USC/Computational Introduction to Deep Learning/Project/Data/Minor/'

filelist_major = os.path.join(path_major, 'Major_1.wav')
filelist_minor = os.path.join(path_minor, 'Minor_1.wav')

fs, signal_major = wavfile.read(filelist_major)
_, signal_minor = wavfile.read(filelist_minor)

sig_len = 94815
time = np.linspace(0., sig_len/fs, sig_len)

freq, time, mag_major = spectrogram(signal_major[:sig_len], fs, nperseg=10000, nfft = 50000)
_, _, mag_minor = spectrogram(signal_minor[:sig_len], fs, nperseg=10000, nfft = 50000)
```

```python
plt.plot(signal_major[:sig_len], label = "Major")
plt.plot(signal_minor[:sig_len], label = "Minor")
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.legend(loc="best")
plt.show()
```

```python
c = plt.pcolormesh(time, freq[:6500], np.log(mag_major[:6500] + 1e-6), cmap = "viridis")
tmp = c.get_clim()
c.set_clim(-10, 10)
plt.axis('off')
plt.show()
```
